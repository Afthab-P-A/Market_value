```python
# iqr method
q1=fifa['overall_rating'].quantile(.25)
q3=fifa['overall_rating'].quantile(.75)
iqr=q3-q1
iqr
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
print(f"lower limit ={lower_bound},upper limit ={upper_bound}")
```

```python
fifa=fifa[(fifa['overall_rating']>=lower_bound) & (fifa['overall_rating']<=upper_bound)]
fifa.shape
```

```python
cor=fifa.corr(numeric_only=True)
cor
```

```python
threshold = 0.7
correlat = set()
columns = cor.columns
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        if abs(cor.iloc[i, j]) >= threshold:
            correlat.add(columns[i])
cor_remove=list(correlat)
cor_remove
```

```python
fifa.drop(cor_remove,axis=1,inplace=True)
```

```python
x=fifa.drop('value_euro',axis=1)
x
```

```python
fifa['value_euro']
```

```python
y=fifa['value_euro']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
cat_pipe=Pipeline(steps=[
    ('cat',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])
```

```python
col_trans=ColumnTransformer(transformers=[
    ('cat',cat_pipe,['positions'])
],remainder='passthrough',n_jobs=-1)
```

```python
x_test.columns
```

```python
x_test.iloc[100].values
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=make_pipeline(col_trans,RandomForestRegressor())
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
mape=mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
r2=r2_score(y_test,y_pred)
r2
```
